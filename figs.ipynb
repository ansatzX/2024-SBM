```python
import os 
import numpy as np 
import pickle
from matplotlib import pyplot as plt
import matplotlib
import shutil
import seaborn as sns
```

```python
mutual_all_lines ={}
for line in sorted([ s for s in sorted(os.listdir()) if s.startswith('traj_s') ]):
    points = { i_step : f"{i_step:04}_step_mutual_infos.pickle" for i_step in range(100) }
    key = f"{line.split('_')[1]}-{line.split('_')[2]}"
    mutual_all_lines[key] = points
```

```python
def read_line(prefix_folder, line_dict):
    mutual_infos = []
    for key in sorted(list(line_dict.keys())):
        step_pickle = line_dict[key]
        
        with open(os.path.join(prefix_folder, step_pickle), 'rb') as f:
            infos = pickle.load(f)
            # vn_entropy_1site
        mutual_infos.append(infos)
    return mutual_infos


def read_omega(prefix_folder):

    with open(os.path.join(prefix_folder, 'sdf_wang1_omega.pickle'), 'rb') as f:
        omegas = pickle.load(f)
        # vn_entropy_1site
    return omegas.tolist()
```
